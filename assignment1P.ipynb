```python
from transformers import pipeline
pipe = pipeline("question-answering",model='deepset/roberta-base-squad2')
```

```python
text = '''Python is a high-level, interpreted programming language known for its simplicity and readability. It was created by Guido van Rossum and first released in 1991. Python supports multiple programming paradigms, including procedural, object-oriented, and functional programming. It is designed to be easy to understand and write, with a syntax that emphasizes readability and reduces the cost of program maintenance.

Python has a dynamic type system and automatic memory management, which makes it an excellent choice for rapid application development. Its extensive standard library and third-party modules allow developers to handle various tasks without having to write a lot of code from scratch. Python is widely used in various domains, including web development, data analysis, artificial intelligence, scientific computing, and automation.

One of Python's key features is its use of indentation to define code blocks, which helps maintain a clean and readable code structure. Python also supports interactive mode, which allows developers to test snippets of code quickly. Popular frameworks and libraries such as Django for web development, Pandas for data manipulation, and TensorFlow for machine learning are built on Python, making it a versatile tool for many programming tasks. '''
myinput = {
    'question': 'Who created Python and in what year was it first released give me name and date?',

    'context':text

}
```

print the output
